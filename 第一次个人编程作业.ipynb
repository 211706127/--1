```python
def panduan(a):#对最开始语句的判断来执行相应的函数
    if (a.find('等于')!=-1):
        chushi(a)
    if (a.find('增加')!=-1):
        jiajian(a)
    if (a.find('看看')!=-1):
        kankan(qiwen)
    if (a.find('减少')!=-1):
        jiajian(a)
```

```python
def chushi(str):#初始化一个全局变量
    global qiwen
    a=str.split(' ')
    if (len(a[3].split('十'))==1):#判定初始时是否含有'十'
        qiwen=shuzi(a[3])
    if(len(a[3].split('十'))==2):
        qiwen=shuzibai(a[3])
```

```python
def jiajian(str):#对加减函数的调用
    global qiwen
    s=str.split(' ')
    qiwen=fuhao(s)
```

```python
def kankan(a):#看看 钱包的输出判断
    if (a>10):
        if(a>10):
            print(hanzibai(qiwen))
        else:
            print(hanzi(qiwen))
        print("看看 “钱太多了”")
    else:
        print(hanzi(qiwen))
        print("看看 “我穷死了”")
```

```python
def fuhao(a):#增加和减少的运算  这里只有百以内的加减
    if (a[1] == '增加'):
        if (len(a[2].split('十'))==1):#判定初始时是否含有'十'
            return qiwen+shuzi(a[2])
        if(len(a[2].split('十'))==2):
            return qiwen+shuzibai(a[2])
    if (a[1] == '减少'):
        if (len(a[2].split('十'))==1):#判定初始时是否含有'十'
            return qiwen-shuzi(a[2])
        if(len(a[2].split('十'))==2):
            return qiwen-shuzibai(a[2])
```

```python
def shuzibai(s):#含十的汉字转文字
    if (len(s)!=3):
         if (s == '十'):
            return 10
         if (s == '二十'):
            return 20
         if (s == '三十'):
            return 30
         if (s == '四十'):
            return 40
         if (s == '五十'):
            return 50
         if (s == '六十'):
            return 60
         if (s == '七十'):
            return 70
         if (s == '八十'):
            return 80
         if (s == '九十'):
            return 90
    else:
        a=s.split('十')
        if (a[0] == '一'):
            z=10
        if (a[0] == '二'):
            z=20
        if (a[0] == '三'):
            z=30
        if (a[0] == '四'):
            z=40
        if (a[0] == '五'):
            z=50
        if (a[0] == '六'):
            z=60
        if (a[0] == '七'):
            z=70
        if (a[0] == '八'):
            z=80
        if (a[0] == '九'):
            z=90
        if (a[1] == '一'):
            x=1
        if (a[1] == '二'):
            x=2
        if (a[1] == '三'):
            x=3
        if (a[1] == '四'):
            x=4
        if (a[1] == '五'):
            x=5
        if (a[1] == '六'):
            x=6
        if (a[1] == '七'):
            x=7
        if (a[1] == '八'):
            x=8
        if (a[1] == '九'):
            x=9
        return z+x
```

```python
def shuzi(a):#不含十的汉字转文字
    if (a == '零'):
        return 0
    if (a == '一'):
        return 1
    if (a == '二'):
        return 2
    if (a == '三'):
        return 3
    if (a == '四'):
        return 4
    if (a == '五'):
        return 5
    if (a == '六'):
        return 6
    if (a == '七'):
        return 7
    if (a == '八'):
        return 8
    if (a == '九'):
        return 9
    if (a == '一百'):
        return 100
```

```python
def hanzi(a):#二十以内数字转汉字
    if (a == 0):
        return '零'
    if (a == 1):
        return '一'
    if (a == 2):
        return '二'
    if (a == 3):
        return '三'
    if (a == 4):
        return '四'
    if (a == 5):
        return '五'
    if (a == 6):
        return '六'
    if (a == 7):
        return '七'
    if (a == 8):
        return '八'
    if (a == 9):
        return '九'
    if (a == 10):
        return '十'
    if (a == 11):
        return '十一'
    if (a == 12):
        return '十二'
    if (a == 13):
        return '十三'
    if (a == 14):
        return '十四'
    if (a == 15):
        return '十五'
    if (a == 16):
        return '十六'
    if (a == 17):
        return '十七'
    if (a == 18):
        return '十八'
    if (a == 19):
        return '十九'
    if (a == 20):
        return '二十'
```

```python
def hanzibai(a):#二十到一百的数字转汉字
    if a%10==0:
        if (a/10 == 10):
            return '一百'
        if (a/10 == 2):
            return '二十'
        if (a/10 == 3):
            return '三十'
        if (a/10 == 4):
            return '四十'
        if (a/10 == 5):
            return '五十'
        if (a/10 == 6):
            return '六十'
        if (a/10 == 7):
            return '七十'
        if (a/10 == 8):
            return '八十'
        if (a/10 == 9):
            return '九十'
    else:
        s=a//10
        d=a%10
        if (s == 2):
            z = '二十'
        if (s == 3):
            z = '三十'
        if (s == 4):
            z = '四十'
        if (s == 5):
            z = '五十'
        if (s == 6):
            z = '六十'
        if (s == 7):
            z = '七十'
        if (s == 8):
            z = '八十'
        if (s == 9):
            z = '九十'
        if (d == 1):
            x = '一'
        if (d == 2):
            x = '二'
        if (d == 3):
            x = '三'
        if (d == 4):
            x = '四'
        if (d == 5):
            x = '五'
        if (d == 6):
            x = '六'
        if (d == 7):
            x = '七'
        if (d == 8):
            x = '八'
        if (d == 9):
            x = '九'
        c=z+x
        return c
```
